```python
import pandas as pd

pakistan_weather = pd.DataFrame({
    "city": ["swabi","maneri","peshawar"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
pakistan_weather
```

```python
india_weather = pd.DataFrame({
    "city": ["Banu","Bunair","Islamabad"],
    "temperature": [38,65,20],
    "humidity": [60, 60, 78]
})
india_weather
```

```python
new_df = pd.concat([pakistan_weather,india_weather])
new_df
```

```python
# so it takes the indexes of separate dataframes in order to stop such thing then use 
new = pd.concat([pakistan_weather,india_weather] , ignore_index=True)
new
```

```python
new_data = pd.concat([pakistan_weather,india_weather] , ignore_index=True)
new_data
```

```python
#series can also be concatenate with dataframes
series = pd.Series(['humidity','sunny','snowy','rainy','hazy','humidity','sunny','snowy','rainy','hazy'] , name='Event')
series
```

```python
pd.concat([pakistan_weather,india_weather,series])
```

```python
# Now what if some one tell us to fill all null values with 0 or some string
df.fillna(0)
```

```python
#fill values 
df.fillna('khalid')
```

```python
# What if we fill values according our own choice then
new_df = df.fillna({
    "temperature":0,
    "windspeed":0,
    'event':"khalid mansoor"
})
new_df
```

```python
# let get a deep dive into method parameter by using method= 'bfill' 
# it will take value nearer or equal to that backward value
df.fillna(method='ffill')

# by using limit = 1 it will only fill one value
```

```python
new = df.fillna(method='bfill' , axis = 0)
new
```

# interpolate(method ='linear') by default method is linear 

```python
new
```

```python
new_df = new.interpolate(method='ffill') #by default the method is linear
new_df
```

```python
# Notice that in above temperature on 2017-01-04 is 29 instead of 30 (in plain linear interpolate)
# There are many other methods for interpolation such as quadratic, piecewise_polynomial, cubic etc.
# Just google "dataframe interpolate" to see complete documentation
```
